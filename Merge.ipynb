```python
import mysql.connector
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="123456789",
    database="Electronics" 
)
cursor = connection.cursor()
cursor.execute("""CREATE TABLE customersales(
    CustomerKey INT,
    Gender VARCHAR(10),
    Name VARCHAR(255),
    City VARCHAR(255),
    State_Code VARCHAR(5),
    State VARCHAR(100),
    Zip_Code VARCHAR(100),
    Country VARCHAR(100),
    Continent VARCHAR(100),
    Birthday DATE,
    Age INT,
    Age_Group VARCHAR(50),
    Order_Number INT,
    Line_Item INT,
    Order_Date DATE,
    Delivery_DATE VARCHAR(200),
    StoreKey INT,
    ProductKey INT,
    Quantity INT,
    Currency_Code VARCHAR(10)
)""")
print("Table Created successfully")
```
